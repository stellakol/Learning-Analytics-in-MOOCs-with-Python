# Analysis for the process of 5 Quizzes                   CLUSTER (START)

```python
import pandas as pd
import numpy as np
import seaborn as sn
import scipy.stats as st
import statsmodels.stats.multicomp as ml
from scipy.stats import f_oneway
from statsmodels.stats.multicomp import pairwise_tukeyhsd
import scikit_posthocs as sp
import matplotlib
import matplotlib.pyplot as plt 
import scipy.stats as stats
%matplotlib inline
```

```python
#ΕΙΣΑΓΩΓΗ ΔΕΔΟΜΕΝΩΝ
df = pd.read_excel('MOOC3.xlsx', sheet_name= "GENERAL", skiprows=1)
df.head()
#ΕΠΙΛΟΓΗ ΣΥΓΚΕΚΡΙΜΕΝΩΝ ΣΤΗΛΩΝ (total_grade,Cluster (START),Cluster (AFTER),Subgroup,QUIZ  1,QUIZ  2,QUIZ  3,QUIZ  4,QUIZ  5)
df2=df[['total_grade', 'Cluster (START)', 'Cluster (AFTER)', 'Subgroup', 'QUIZ  1', 'QUIZ  2', 'QUIZ  3', 'QUIZ  4', 'QUIZ  5']]
#ΕΠΙΛΟΓΗ SUBGROUP ΝΑ ΕΙΝΑΙ ΙΣΣΟ ΜΕ 1 
#one= df.Subgroup==1
#df_by_one= df[one]
df2
```

```python
#EΛΕΓΧΟΣ ΓΙΑ ΚΑΘΕ ΜΑΘΗΤΗ ΑΝ ΕΧΕΙ ΑΠΑΝΤΗΣΕΙ ΣΕ ΤΟΥΛΑΧΙΣΤΟΝ 2 QUIZ ΣΥΝΟΛΙΚΑ. ΑΝ ΕΧΕΙ <2 AΠΑΝΤΗΜΕΝΑ QUIZ ΤΟ ΚΑΝΩ DROP 

Quiz_list=["QUIZ  1", "QUIZ  2","QUIZ  3", "QUIZ  4", "QUIZ  5"]


for i in range(df2.shape[0]):
    counter = 0
    if df2.at[i,'Subgroup'] == 1:
        
        for j in Quiz_list:
            
            if df2.at[i,j]!="-":             
                counter+=1                       
        if counter<2:
            
            df2.drop(df.index[i], inplace=True)           
    else:
        df2.drop(df.index[i], inplace=True)
        
          
df2
```

```python
df2quiz=df2[["QUIZ  1", "QUIZ  2","QUIZ  3", "QUIZ  4", "QUIZ  5"]]
df2quiz.replace(to_replace='-',value=float('nan'), inplace=True)
df2quiz['mean'] = df2quiz.mean(axis=1, skipna=True)

df2info=df2[['total_grade', 'Cluster (START)', 'Cluster (AFTER)', 'Subgroup']]

result= [df2info, df2quiz]
dfquiz_total= pd.concat(result, axis=1)

dfquiz_total
```

```python
#Cluster Start 1 (LOW>4)
cs1=dfquiz_total.groupby('Cluster (START)').get_group(1)

cs1
col=cs1.columns
low=cs1[col[9:10:1]]
low.dropna(inplace=True)

meanlow = low.mean(axis=0)#μεσος ορος βαθμολογίας μαθητων low στa 5 quizzes

low_list=[]
for i, row in low.iterrows():
    low_list.append(low.at[i,'mean'])
low_list
```

```python
#Cluster Start 2 (MEDIUM 4-5)
cs2=dfquiz_total.groupby('Cluster (START)').get_group(2)

cs2
col=cs2.columns
medium=cs2[col[9:10:1]]
medium.dropna(inplace=True)

meanmedium = medium.mean(axis=0)#μεσος ορος βαθμολογίας μαθητων medium στa 5 quizzes

medium_list=[]
for i, row in medium.iterrows():
    medium_list.append(medium.at[i,'mean'])
medium_list
```

```python
#Cluster Start 3 (LARGE 5-6)
cs3=dfquiz_total.groupby('Cluster (START)').get_group(3)

cs3
col=cs3.columns
large=cs3[col[9:10:1]]
large.dropna(inplace=True)

meanlarge = large.mean(axis=0)#μεσος ορος βαθμολογίας μαθητων large στa 5 quizzes

large_list=[]
for i, row in large.iterrows():
    large_list.append(large.at[i,'mean'])
large_list
```

```python
#Cluster Start 4 (XLARGE <6)
cs4=dfquiz_total.groupby('Cluster (START)').get_group(4)

cs4
col=cs4.columns
xlarge=cs4[col[9:10:1]]
xlarge.dropna(inplace=True)

meanxlarge = xlarge.mean(axis=0)#μεσος ορος βαθμολογίας μαθητων xlarge στa 5 quizzes

xlarge_list=[]
for i, row in xlarge.iterrows():
    xlarge_list.append(xlarge.at[i,'mean'])
xlarge_list
```

# one way ANOVA test

```python
# Shapiro-Wilk normality test
stats.shapiro(low_list), stats.shapiro(medium_list), stats.shapiro(large_list), stats.shapiro(xlarge_list)

#Όλες οι τιμές p εκτός από την ομάδα medium είναι μικρότερες από το όριο a = 0,05, 
#επομένως απορρίπτεται η μηδενική υπόθεση 
#τα δείγματα προέρχονται από πληθυσμούς που δεν ακολουθούν την κανονική κατανομή
```

```python
# Levene variance test  
stats.levene(low_list, medium_list, large_list, xlarge_list)

#Η τιμή p είναι μικρότερη από το όριο a = 0,05, 
#επομένως απορρίπτεται η μηδενική υπόθεση 
#τα δείγματα προέρχονται από πληθυσμούς με διαφορετική διακύμανση
```

```python
#Apply one way ANOVA by calling the f_oneway() method
F, p = stats.f_oneway(low_list, medium_list, large_list, xlarge_list)
print('F statistic = {:5.3f} and probability p = {:5.3f}'.format(F, p)) 
```

# Kruskal Wallis test

```python
stats.kruskal(low_list, medium_list, large_list, xlarge_list)
```

# Dunn’s test

```python
#perform Dunn's test using a Bonferonni correction for the p-values

data = [low_list, medium_list, large_list, xlarge_list]
data

sp.posthoc_dunn(data , p_adjust = 'bonferroni')
```

```python
from numpy import mean
from numpy import var
from math import sqrt

n1,n2= len(low_list), len(medium_list)
s1,s2= var(low_list, ddof=1), var(medium_list, ddof=1)
s= sqrt(((n1-1)*s1+(n2-1)*s2)/(n1+n2-2))
u1,u2= mean(low_list), mean(medium_list)
d= (u1-u2)/s

cohen= abs(d)
if cohen<0.10:
    print('Very Small Effect Size=',cohen)
elif cohen<0.1:
    print('Small Effect Size=',cohen)
elif cohen<0.50:
    print('Medium Effect Size=',cohen)
elif cohen<0.80:
    print('Large Effect Size=',cohen)
else:
    print('Very Large=', cohen)
```
