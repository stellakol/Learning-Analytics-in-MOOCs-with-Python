# Analysis for the process of 5 exercises                   CLUSTER (START)

```python
import pandas as pd
import numpy as np
import seaborn as sn
import scipy.stats as st
import statsmodels.stats.multicomp as ml
from scipy.stats import f_oneway
from statsmodels.stats.multicomp import pairwise_tukeyhsd
import scikit_posthocs as sp
import matplotlib
import matplotlib.pyplot as plt 
import scipy.stats as stats
%matplotlib inline
```

```python
#EΙΣΑΓΩΓΗ ΔΕΔΟΜΕΝΩΝ
df = pd.read_excel('MOOC3.xlsx', sheet_name= "GENERAL", skiprows=1)
df.head()
#ΕΠΙΛΟΓΗ ΣΥΓΚΕΚΡΙΜΕΝΩΝ ΣΤΗΛΩΝ (total_grade,Cluster (START),Cluster (AFTER),Subgroup,QUIZ  1,QUIZ  2,QUIZ  3,QUIZ  4,QUIZ  5)
df4=df[['total_grade', 'Cluster (START)', 'Cluster (AFTER)', 'Subgroup', 'Άσκηση  1','Άσκηση 2','Άσκηση 3','Άσκηση 4','Άσκηση5']]
#ΕΠΙΛΟΓΗ SUBGROUP ΝΑ ΕΙΝΑΙ ΙΣΣΟ ΜΕ 1 
#one= df.Subgroup==1
#df_by_one= df[one]
df4
```

```python
#EΛΕΓΧΟΣ ΓΙΑ ΚΑΘΕ ΜΑΘΗΤΗ ΑΝ ΕΧΕΙ ΑΠΑΝΤΗΣΕΙ ΣΕ ΤΟΥΛΑΧΙΣΤΟΝ 2 ΑΣΚΗΣΕΙΣ ΣΥΝΟΛΙΚΑ. ΑΝ ΕΧΕΙ <2 AΠΑΝΤΗΜΕΝΑ QUIZ ΤΟ ΚΑΝΩ DROP 

exercise_list=['Άσκηση  1','Άσκηση 2','Άσκηση 3','Άσκηση 4','Άσκηση5']


for i in range(df4.shape[0]):
    counter = 0
    if df4.at[i,'Subgroup'] == 1:
        
        for j in exercise_list:
            
            if df4.at[i,j]!="-":             
                counter+=1                       
        if counter<2:
            
            df4.drop(df.index[i], inplace=True)           
    else:
        df4.drop(df.index[i], inplace=True)
        
          
df4
```

```python
df4ex=df4[['Άσκηση  1','Άσκηση 2','Άσκηση 3','Άσκηση 4','Άσκηση5']]
df4ex.replace(to_replace='-',value=float('nan'), inplace=True)
df4ex['mean'] = df4ex.mean(axis=1, skipna=True)

df4info=df4[['total_grade', 'Cluster (START)', 'Subgroup']]
#dfclu=df[df.columns[228:229:1]]

result= [df4info, df4ex]
dfex_total= pd.concat(result, axis=1)

dfex_total
```

```python
#Cluster Start 1 (LOW>4)
cs1ex=dfex_total.groupby('Cluster (START)').get_group(1)

cs1ex
col=cs1ex.columns
exlow=cs1ex[col[8:9:1]]
exlow.dropna(inplace=True)#κρατάμε μόνο την στήλη των μέσων όρων

meanexlow = exlow.mean(axis=0)#μεσος ορος βαθμολογίας μαθητων low στις 5 ασκήσεις

exlow_list=[]
for i, row in exlow.iterrows():
    exlow_list.append(exlow.at[i,'mean'])
exlow_list
```

```python
#Cluster Start 2 (MEDIUM 4-5)
cs2ex=dfex_total.groupby('Cluster (START)').get_group(2)

cs2ex
col=cs2ex.columns
exmedium=cs2ex[col[8:9:1]]
exmedium.dropna(inplace=True)#κρατάμε μόνο την στήλη των μέσων όρων

meanexmedium = exmedium.mean(axis=0)#μεσος ορος βαθμολογίας μαθητων medium στις 5 ασκήσεις

exmedium_list=[]
for i, row in exmedium.iterrows():
    exmedium_list.append(exmedium.at[i,'mean'])
exmedium_list
```

```python
#Cluster Start 3 (LARGE 5-6)
cs3ex=dfex_total.groupby('Cluster (START)').get_group(3)

cs3ex
col=cs3ex.columns
exlarge=cs3ex[col[8:9:1]]
exlarge.dropna(inplace=True)#κρατάμε μόνο την στήλη των μέσων όρων

meanexlarge = exlarge.mean(axis=0)#μεσος ορος βαθμολογίας μαθητων large στις 5 ασκήσεις

exlarge_list=[]
for i, row in exlarge.iterrows():
    exlarge_list.append(exlarge.at[i,'mean'])
exlarge_list
```

```python
#Cluster Start 4 (XLARGE <6)
cs4ex=dfex_total.groupby('Cluster (START)').get_group(4)

cs4ex
col=cs4ex.columns
ex_xlarge=cs4ex[col[8:9:1]]
ex_xlarge.dropna(inplace=True)#κρατάμε μόνο την στήλη των μέσων όρων

meanex_xlarge = ex_xlarge.mean(axis=0)#μεσος ορος βαθμολογίας μαθητων xlarge στις 5 ασκήσεις

ex_xlarge_list=[]
for i, row in ex_xlarge.iterrows():
    ex_xlarge_list.append(ex_xlarge.at[i,'mean'])
ex_xlarge_list
```

# one way ANOVA test

```python
# Shapiro-Wilk normality test
stats.shapiro(exlow_list), stats.shapiro(exmedium_list), stats.shapiro(exlarge_list), stats.shapiro(ex_xlarge_list)

#Όλες οι τιμές p είναι μικρότερες από το όριο a = 0,05, 
#επομένως δεν «αποτυγχάνουμε να απορρίψουμε» τη μηδενική υπόθεση 
#τα δείγματα προέρχονται από πληθυσμούς που δεν ακολουθούν την κανονική κατανομή
```

```python
# Levene variance test  
stats.levene(exlow_list, exmedium_list, exlarge_list, ex_xlarge_list)

#Η τιμή p είναι μεγαλύτερη από το όριο a = 0,05, 
#επομένως «αποτυγχάνουμε να απορρίψουμε» τη μηδενική υπόθεση 
#τα δείγματα προέρχονται από πληθυσμούς με την ίδια διακύμανση
```

```python
#Apply one way ANOVA by calling the f_oneway() method
F, p = stats.f_oneway(exlow_list, exmedium_list, exlarge_list,ex_xlarge_list)
print('F statistic = {:5.3f} and probability p = {:5.3f}'.format(F, p))
```


# Kruskal Wallis test

```python
stats.kruskal(exlow_list, exmedium_list, exlarge_list, ex_xlarge_list)
```

# Dunn’s test

```python
#perform Dunn's test using a Bonferonni correction for the p-values

data = [exlow_list, exmedium_list, exlarge_list, ex_xlarge_list]
data

sp.posthoc_dunn(data , p_adjust = 'bonferroni')
```

# Cohen's Effect Size

```python
from numpy import mean
from numpy import var
from math import sqrt

n1,n2= len(exlow_list), len(exmedium_list)
s1,s2= var(exlow_list, ddof=1), var(exmedium_list, ddof=1)
s= sqrt(((n1-1)*s1+(n2-1)*s2)/(n1+n2-2))
u1,u2= mean(exlow_list), mean(exmedium_list)
d= (u1-u2)/s

cohen= abs(d)
if cohen<0.10:
    print('Very Small Effect Size=',cohen)
elif cohen<0.1:
    print('Small Effect Size=',cohen)
elif cohen<0.50:
    print('Medium Effect Size=',cohen)
elif cohen<0.80:
    print('Large Effect Size=',cohen)
else:
    print('Very Large=', cohen)

    
```
