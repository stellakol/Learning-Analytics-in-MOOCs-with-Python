```python
#ΕΙΣΑΓΩΓΗ ΔΕΔΟΜΕΝΩΝ
df = pd.read_excel('MOOC3.xlsx', sheet_name= "GENERAL", skiprows=1)
```

```python
#ΕΞΑΓΩΓΗ ΤΩΝ ΣΤΗΛΩΝ ΜΕ ΤΑ ΕΡΩΤΗΜΑΤΑ ΚΑΙ Μ.Ο ΒΑΥΜΟΛΟΓΙΑΣ ΚΑΘΕ ΦΟΙΤΗΤΗ

dfinfo=df[[ 'Cluster (START)',  'Subgroup']]
dfbot1a=df[df.columns[127:128:1]]
dfbot1b=df[df.columns[129:133:1]]

dfbot2a=df[df.columns[135:136:1]]
dfbot2b=df[df.columns[137:141:1]]

dfbot3a=df[df.columns[143:144:1]]
dfbot3b=df[df.columns[145:149:1]]

dfbot4a=df[df.columns[151:152:1]]
dfbot4b=df[df.columns[153:157:1]]

dfbot5a=df[df.columns[159:160:1]]
dfbot5b=df[df.columns[161:165:1]]


result = [dfinfo, dfbot1a,dfbot1b, dfbot2a, dfbot2b, dfbot3a, dfbot3b, dfbot4a, dfbot4b, dfbot5a, dfbot5b]

  
dfbot_total= pd.concat(result, axis=1)
dfbot_total
```

```python
#EΛΕΓΧΟΣ ΓΙΑ ΚΑΘΕ ΜΑΘΗΤΗ ΑΝ ΕΧΕΙ ΑΠΑΝΤΗΣΕΙ ΣΕ ΤΟΥΛΑΧΙΣΤΟΝ 2 CHAT ΣΥΝΟΛΙΚΑ. ΑΝ ΕΧΕΙ <2 AΠΑΝΤΗΜΕΝΑ CHAT ΤΟ ΚΑΝΩ DROP 

dfbot_total.replace(to_replace=float('nan'),value='-', inplace=True)
for i in range(dfbot_total.shape[0]):
    counter = 0
    if dfbot_total.at[i,'Subgroup'] == 1:
        
        for j in range(2,27,1):
            
            if dfbot_total.at[i,dfbot_total.columns[j]]!="-":             
                counter+=1                       
        if counter<=10:
            
            dfbot_total.drop(df.index[i], inplace=True)           
    else:
        dfbot_total.drop(df.index[i], inplace=True)
        
          
dfbot_total.replace(to_replace='-',value=float('nan'), inplace=True)
dfbot_total
```

```python
#ΔΗΜΙΟΥΡΓΙΑ 5 ΝΕΑ DF ΜΕ ΤΗΝ ΚΑΘΕ ΕΡΩΤΗΣΗ ΚΑΙ ΓΙΑ ΤΙΣ 5 ΕΒΔΟΜΑΔΕΣ ΣΥΝΟΛΙΚΟ ΠΛΗΘΟΣ 523

#Στη συζήτηση που είχα στο chat νομίζω ότι συνεργάστηκα καλά με τον/την συνεργάτη μου.
dfinfo=dfbot_total[[ 'Cluster (START)',  'Subgroup']]
dfchat1a=dfbot_total[dfbot_total.columns[2:3:1]]
dfchat1b=dfbot_total[dfbot_total.columns[3:4:1]]
dfchat1c=dfbot_total[dfbot_total.columns[4:5:1]]
dfchat1d=dfbot_total[dfbot_total.columns[5:6:1]]
dfchat1e=dfbot_total[dfbot_total.columns[6:7:1]]

#Μού άρεσε που υπήρχε ο πράκτορας και έκανε παρεμβάσεις στη συζήτηση.
dfchat2a=dfbot_total[dfbot_total.columns[7:8:1]]
dfchat2b=dfbot_total[dfbot_total.columns[8:9:1]]
dfchat2c=dfbot_total[dfbot_total.columns[9:10:1]]
dfchat2d=dfbot_total[dfbot_total.columns[10:11:1]]
dfchat2e=dfbot_total[dfbot_total.columns[11:12:1]]

#Νομίζω ότι οι παρεμβάσεις του πράκτορα ήταν σχετικές με το τι συζητούσαμε εκείνη τη στιγμή.
dfchat3a=dfbot_total[dfbot_total.columns[12:13:1]]
dfchat3b=dfbot_total[dfbot_total.columns[13:14:1]]
dfchat3c=dfbot_total[dfbot_total.columns[14:15:1]]
dfchat3d=dfbot_total[dfbot_total.columns[15:16:1]]
dfchat3e=dfbot_total[dfbot_total.columns[16:17:1]]

#Νομίζω ότι οι παρεμβάσεις του πράκτορα με έκαναν να σκεφτώ καλύτερα γι αυτό που μαθαίνω.
dfchat4a=dfbot_total[dfbot_total.columns[17:18:1]]
dfchat4b=dfbot_total[dfbot_total.columns[18:19:1]]
dfchat4c=dfbot_total[dfbot_total.columns[19:20:1]]
dfchat4d=dfbot_total[dfbot_total.columns[20:21:1]]
dfchat4e=dfbot_total[dfbot_total.columns[21:22:1]]

#Με ενόχλησαν οι παρεμβάσεις του πράκτορα καθώς συζητούσα με τον/την συνεργάτη μου.
dfchat5a=dfbot_total[dfbot_total.columns[22:23:1]]
dfchat5b=dfbot_total[dfbot_total.columns[23:24:1]]
dfchat5c=dfbot_total[dfbot_total.columns[24:25:1]]
dfchat5d=dfbot_total[dfbot_total.columns[25:26:1]]
dfchat5e=dfbot_total[dfbot_total.columns[26:27:1]]


result1 = [dfchat1a, dfchat2a, dfchat3a, dfchat4a, dfchat5a]
dfchat1= pd.concat(result1, axis=1)
dfchat1


result2 = [dfchat1b, dfchat2b, dfchat3b, dfchat4b, dfchat5b]
dfchat2= pd.concat(result2, axis=1)
dfchat2


result3 = [dfchat1c, dfchat2c, dfchat3c, dfchat4c, dfchat5c]
dfchat3= pd.concat(result3, axis=1)
dfchat3


result4 = [dfchat1d, dfchat2d, dfchat3d, dfchat4d, dfchat5d]
dfchat4= pd.concat(result4, axis=1)
dfchat4


result5 = [dfchat1e, dfchat2e, dfchat3e, dfchat4e, dfchat5e]
dfchat5= pd.concat(result5, axis=1)
dfchat5
```

# CLUSTER(START) QUESTION 1 
#Στη συζήτηση που είχα στο chat νομίζω ότι συνεργάστηκα καλά με τον/την συνεργάτη μου.

```python
#Cluster Start 1 (LOW>4)
q1_1=dfquest1.groupby('Cluster (START)').get_group(1)


col=q1_1.columns
q1low=q1_1[col[2:3:1]]#κρατάμε μόνο την στήλη των μέσων όρων
q1low.dropna(inplace=True)

meanq1low = q1low.mean(axis=0)#μεσος ορος μαθητων low στην ερώτηση 1


q1low_list=[]
for i, row in q1low.iterrows():
    q1low_list.append(q1low.at[i,'mean'])


q1low_list
```

```python
#Cluster Start 2 (MEDIUM 4-5)
q1_2=dfquest1.groupby('Cluster (START)').get_group(2)


col=q1_2.columns
q1medium=q1_2[col[2:3:1]]#κρατάμε μόνο την στήλη των μέσων όρων
q1medium.dropna(inplace=True)

meanq1medium = q1medium.mean(axis=0)#μεσος ορος μαθητων Medium στην ερώτηση 1


q1medium_list=[]
for i, row in q1medium.iterrows():
    q1medium_list.append(q1medium.at[i,'mean'])


q1medium_list
```

```python
#Cluster Start 3 (LARGE 5-6)
q1_3=dfquest1.groupby('Cluster (START)').get_group(3)


col=q1_3.columns
q1large=q1_3[col[2:3:1]]#κρατάμε μόνο την στήλη των μέσων όρων
q1large.dropna(inplace=True)

meanq1large = q1large.mean(axis=0)#μεσος ορος μαθητων Large στην ερώτηση 1


q1large_list=[]
for i, row in q1large.iterrows():
    q1large_list.append(q1large.at[i,'mean'])


q1large_list
```

```python
#Cluster Start 4 (XLARGE <6)
q1_4=dfquest1.groupby('Cluster (START)').get_group(4)


col=q1_4.columns
q1xlarge=q1_4[col[2:3:1]]#κρατάμε μόνο την στήλη των μέσων όρων
q1xlarge.dropna(inplace=True)

meanq1xlarge = q1xlarge.mean(axis=0)#μεσος ορος μαθητων XLarge στην ερώτηση 1


q1xlarge_list=[]
for i, row in q1xlarge.iterrows():
    q1xlarge_list.append(q1xlarge.at[i,'mean'])


q1xlarge_list
```

# one way ANOVA test

```python
# Shapiro-Wilk normality test
stats.shapiro(q1low_list), stats.shapiro(q1medium_list), stats.shapiro(q1large_list), stats.shapiro(q1xlarge_list)

#Όλες οι τιμές p είναι μικρότερες από το όριο a = 0,05, 
#επομένως δεν «αποτυγχάνουμε να απορρίψουμε» τη μηδενική υπόθεση 
#τα δείγματα προέρχονται από πληθυσμούς που δεν ακολουθούν την κανονική κατανομή
```

```python
# Levene variance test  
stats.levene(q1low_list, q1medium_list, q1large_list, q1xlarge_list)

#Η τιμή p είναι μικρότερα από το όριο a = 0,05, 
#επομένως δεν «αποτυγχάνουμε να απορρίψουμε» τη μηδενική υπόθεση 
#τα δείγματα προέρχονται από πληθυσμούς με διαφορετική διακύμανση
```

```python
#Apply one way ANOVA by calling the f_oneway() method
F, p = stats.f_oneway(q1low_list, q1medium_list, q1large_list, q1xlarge_list)
print('F statistic = {:5.3f} and probability p = {:5.3f}'.format(F, p)) 
```

# Kruskal Wallis test

```python
stats.kruskal(q1low_list, q1medium_list, q1large_list, q1xlarge_list)
```

# Dunn's test

```python
#perform Dunn's test using a Bonferonni correction for the p-values

data = [q1low_list, q1medium_list, q1large_list, q1xlarge_list]
data

sp.posthoc_dunn(data , p_adjust = 'bonferroni')
```

# Cohen's Effect Size

```python
from numpy import mean
from numpy import var
from math import sqrt

n1,n2= len(q5low_list), len(q5medium_list)
s1,s2= var(q5low_list, ddof=1), var(q5medium_list, ddof=1)
s= sqrt(((n1-1)*s1+(n2-1)*s2)/(n1+n2-2))
u1,u2= mean(q5low_list), mean(q5medium_list)
d= (u1-u2)/s

cohen= abs(d)
if cohen<0.10:
    print('Very Small Effect Size=',cohen)
elif cohen<0.1:
    print('Small Effect Size=',cohen)
elif cohen<0.50:
    print('Medium Effect Size=',cohen)
elif cohen<0.80:
    print('Large Effect Size=',cohen)
else:
    print('Very Large=', cohen)
```
